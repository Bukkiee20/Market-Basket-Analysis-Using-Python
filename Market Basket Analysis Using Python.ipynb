### Step 3: Market Basket Analysis

#### Data preprocessing

```python
csv["Member_number"] = csv["Member_number"].astype(int)
csv["itemDescription"] = csv["itemDescription"].astype(str)
```

### Step 4: Data Visualization
#### Bar plot of top 50 most purchased items

```python
import matplotlib as plt
import seaborn as sns
```

```python
top_items = csv["itemDescription"].value_counts().head(20)
plt.figure(figsize= (10, 6))
sns.barplot(x= top_items.values, y= top_items.index, palette = "viridis")
plt.xlabel("Frequency")
plt.ylabel("Item Description")
plt.title("Top 50 Most Purchased Items")
plt.show()
```
